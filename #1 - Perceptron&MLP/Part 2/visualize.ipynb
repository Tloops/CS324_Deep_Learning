# Visualizer of MLP

### 1. import necessary packages

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

### 2. read json file
The accuracy are recorded during training.

```python
json_file_path = 'accuracy.json'
json_file = open(json_file_path, mode='r')
accuracy = json.load(json_file)

mode = accuracy["mode"]
epoch = np.array(accuracy["epoch"])
acc_train = np.array(accuracy["acc_train"])
acc_test = np.array(accuracy["acc_test"])
```

### 3. plot the curve

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(epoch, acc_train, label="train")
ax.plot(epoch, acc_test, label="test")
ax.set_xlabel("epoch")  # 添加横轴标签
ax.set_ylabel("accuracy(%)")  # 添加纵轴标签
plt.savefig('./img/accuracy_curve_' + mode + '.jpg')
ax.legend(loc="best")
# plt.show()
```
