# Visualizer of MLP

### 1. import necessary packages

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

### 2. read json file
The accuracy are recorded during training.

```python
json_file_path = 'torch_accuracy.json'
json_file = open(json_file_path, mode='r')
accuracy = json.load(json_file)

epoch = np.array(accuracy["step"])
acc_train = np.array(accuracy["acc_train"]) / 100
acc_test = np.array(accuracy["acc_test"]) / 100
```

### 3. plot the curve

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(epoch, acc_train, label="train")
ax.plot(epoch, acc_test, label="test")
ax.set_xlabel("step")  # 添加横轴标签
ax.set_ylabel("accuracy(%)")  # 添加纵轴标签
ax.legend(loc="best")
plt.savefig('./img/accuracy_curve.jpg')
# plt.show()
```
