# Visualizer of MLP

### 1. import necessary packages

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

### 2. plot the numpy version mlp accuracy curve
The accuracy are recorded during training.
Only the first 50 epochs are shown to make the curve clearer.

```python
json_file_path = 'numpy_accuracy.json'
json_file = open(json_file_path, mode='r')
accuracy = json.load(json_file)

mode = accuracy["mode"]
epoch = np.array(accuracy["epoch"])[:50]
acc_train = np.array(accuracy["acc_train"])[:50]
acc_test = np.array(accuracy["acc_test"])[:50]

fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(epoch, acc_train, label="train")
ax.plot(epoch, acc_test, label="test")
ax.set_xlabel("epoch")  # 添加横轴标签
ax.set_ylabel("accuracy(%)")  # 添加纵轴标签
ax.legend(loc="best")
plt.savefig('./img/accuracy_curve_numpy.jpg')
```
