# Visualizer of MLP

### 1. import necessary packages

```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

### 2. read json file
The accuracy are recorded during training.

```python
json_file_path5 = 'accuracy_5.json'
json_file5 = open(json_file_path5, mode='r')
dict5 = json.load(json_file5)

epoch5 = np.array(dict5["step"])
accuracy5 = np.array(dict5["accuracy"])
loss5 = np.array(dict5["loss"])

json_file_path10 = 'accuracy_10.json'
json_file10 = open(json_file_path10, mode='r')
dict10 = json.load(json_file10)

epoch10 = np.array(dict10["step"])
accuracy10 = np.array(dict10["accuracy"])
loss10 = np.array(dict10["loss"])

json_file_path15 = 'accuracy_15.json'
json_file15 = open(json_file_path15, mode='r')
dict15 = json.load(json_file15)

epoch15 = np.array(dict15["step"])
accuracy15 = np.array(dict15["accuracy"])
loss15 = np.array(dict15["loss"])

json_file_path20 = 'accuracy_20.json'
json_file20 = open(json_file_path20, mode='r')
dict20 = json.load(json_file20)

epoch20 = np.array(dict20["step"])
accuracy20 = np.array(dict20["accuracy"])
loss20 = np.array(dict20["loss"])
```

### 3. plot the curve

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(epoch5, accuracy5, 'r-', label="length 5")
ax.plot(epoch10, accuracy10, 'g-', label="length 10")
ax.plot(epoch15, accuracy15, 'b-', label="length 15")
ax.plot(epoch20, accuracy20, 'y-', label="length 20")
ax.set_xlabel("step")  # 添加横轴标签
ax.set_ylabel("accuracy")  # 添加纵轴标签
ax.legend(loc="best")
plt.savefig('./img/accuracy_curve.jpg')
# plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(epoch5, loss5, 'r-', label="length 5")
ax.plot(epoch10, loss10, 'g-', label="length 10")
ax.plot(epoch15, loss15, 'b-', label="length 15")
ax.plot(epoch20, loss20, 'y-', label="length 20")
ax.set_xlabel("step")  # 添加横轴标签
ax.set_ylabel("loss")  # 添加纵轴标签
ax.legend(loc="best")
plt.savefig('./img/loss_curve.jpg')
# plt.show()
```
